```python
crime["total"] = crime["murder"] + crime["manslaughter"] + crime["forcible"] + crime["nonforce"] + crime["robbery"] + crime["agg_aslt"] + crime["burglary"] + crime["vehicle"] + crime["arson"]
crime


```

```python
g = sns.FacetGrid(crime,
                  palette=pinellas_palette,
                  size=7)

g.map(plt.scatter, "total_enr", "burglary", 
      alpha=.8)
```

```python
g = sns.FacetGrid(crime,
                  palette=pinellas_palette,
                  size=7)

g.map(plt.scatter, "total_enr", "robbery", 
      alpha=.8)
```

```python
g = sns.FacetGrid(crime,
                  palette=pinellas_palette,
                  size=7)

g.map(plt.scatter, "total_enr", "agg_aslt", 
      alpha=.8)
```

```python
g = sns.FacetGrid(crime,
                  palette=pinellas_palette,
                  size=7)
g.map(plt.scatter, "arson", "total_enr", 
      alpha=.8, 
      s=60)
```

```python
 g = sns.lmplot(data=crime, x='total', y='total_enr',
             
               fit_reg=False,
               size=7)
sns.regplot(x="total", y="total_enr", data=crime, scatter=False, ax=g.axes[0, 0])                                 
```
